```python
import re
import json

def convert_ts_to_json(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as f:
        data = f.read()

    # 1. Isolate the main dictionary block
    data = data[data.find('{'):data.rfind('}')+1]

    # 2. Remove inline comments
    data = re.sub(r'//.*', '', data)

    # 3. Convert single-quoted strings to double-quoted strings 
    data = re.sub(r"'([^']+)'", r'"\1"', data)

    # 4. Wrap unquoted keys in double quotes
    data = re.sub(r'([a-zA-Z0-9_]+)\s*:', r'"\1":', data)

    # 5. Remove trailing commas before brackets/braces
    data = re.sub(r',\s*\}', '}', data)
    data = re.sub(r',\s*\]', ']', data)

    # 6. Parse and save to a properly formatted JSON file
    try:
        parsed_data = json.loads(data)
        with open(output_file, 'w', encoding='utf-8') as f:
            json.dump(parsed_data, f, indent=4)
        print(f"Successfully generated {output_file}")
    except json.JSONDecodeError as e:
        print(f"Formatting error: {e}")

    return data

```
