```python
import os

folder_path = "gen9-randombattle"
```

```python
for filename in os.listdir(folder_path):
    if filename.endswith(".json"):
        file_path = os.path.join(folder_path, filename)
        b = battle.Battle(file_path)
        row = {}
        for player in [1,2]:
            i = 1
            for key,value in b.teams_full[player-1].items():
                prefix = f'p{player}_mon{i}_'

                for stat in value['stats']:
                    row[prefix + stat] = [value['stats'][stat]]
                row[prefix + 'types'] = [value['types']] if type(value['types']) == list else [[value['types']]]
                i+=1
        row['winner'] = [1] if b.winner == b.p1 else [2]
        df = pd.concat([df, pd.DataFrame(row)], ignore_index=True)
```
