```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=["winner"])
y = df["winner"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
# check how many pokemon have a 3x type advantage against the opposing team

def check_triple_advantage(row):
    p1_teams = [row[f"p1_mon{i}_types"] for i in range(1, 7)]
    p2_teams = [row[f"p2_mon{i}_types"] for i in range(1, 7)]

    def team_has_counter(attacker_team, defender_team):
        for attacker in attacker_team:
            matchup_count = 0
            for defender in defender_team:
                # Check if any attacker type is super effective against any defender type
                if any(
                    modifiers.get(a, {}).get(d, 1.0) > 1.0
                    for a in attacker
                    for d in defender
                ):
                    matchup_count += 1

            if matchup_count >= 3:
                return True
        return False

    # Returns True if either Player 1 or Player 2 has such a Pokémon
    return team_has_counter(p1_teams, p2_teams) or team_has_counter(
        p2_teams, p1_teams
    )


# Calculate frequencies
matching_games = X_train.apply(check_triple_advantage, axis=1)
total_matches = matching_games.sum()
percentage = (total_matches / len(X_train)) * 100

print(
    f"Games in X_train with a 3+ Pokémon type advantage: {total_matches} / {len(X_train)} ({percentage:.2f}%)"
)
```

```python
def count_3x_adv_pokemon(row, attacker_prefix, defender_prefix):
    attacker_team = [row[f"{attacker_prefix}_mon{i}_types"] for i in range(1, 7)]
    defender_team = [row[f"{defender_prefix}_mon{i}_types"] for i in range(1, 7)]

    advantage_count = 0
    for attacker in attacker_team:
        matchups = 0
        for defender in defender_team:
            if any(
                modifiers.get(a, {}).get(d, 1.0) > 1.0
                for a in attacker
                for d in defender
            ):
                matchups += 1
        if matchups >= 3:
            advantage_count += 1

    return advantage_count


# Count for each team per row
p1_counts = X_train.apply(lambda r: count_3x_adv_pokemon(r, "p1", "p2"), axis=1)
p2_counts = X_train.apply(lambda r: count_3x_adv_pokemon(r, "p2", "p1"), axis=1)

# Subtract p1 count from p2 count
X_train["type_adv_diff"] = p2_counts - p1_counts

# Calculate Pearson correlation with y_train
correlation = X_train["type_adv_diff"].corr(y_train)
print(f"Correlation with winner: {correlation:.4f}")
```

```python
p1_hp_cols = [f"p1_mon{i}_hp" for i in range(1, 7)]
p2_hp_cols = [f"p2_mon{i}_hp" for i in range(1, 7)]

# Calculate the total HP differential (Team 1 - Team 2)
X_train["hp_sum_diff"] =  X_train[p2_hp_cols].sum(axis=1) - X_train[p1_hp_cols].sum(axis=1)

# Correlation check
correlation = X_train["hp_sum_diff"].corr(y_train)
print(f"HP Sum Differential Correlation: {correlation:.4f}")
```

```python
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, cross_val_score

# Subset the features
X_cv = X_train[["hp_sum_diff", "type_adv_diff"]]

# Define baseline models
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
}

# 5-Fold Stratified Cross-Validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for name, model in models.items():
    scores = cross_val_score(model, X_cv, y_train, cv=cv, scoring="accuracy")
    print(f"{name:20} | Mean Accuracy: {scores.mean():.4f} (±{scores.std():.4f})")
```

```python
def count_3x_spe_pokemon(row, attacker_prefix, defender_prefix):
    attacker_spe = [row[f"{attacker_prefix}_mon{i}_spe"] for i in range(1, 7)]
    defender_spe = [row[f"{defender_prefix}_mon{i}_spe"] for i in range(1, 7)]

    adv_count = 0
    for s1 in attacker_spe:
        # Count how many opponent pokemon this pokemon outspeeds
        if sum(1 for s2 in defender_spe if s1 > s2) >= 3:
            adv_count += 1
    return adv_count


# Calculate speed tiers and the differential
p1_spe_counts = X_train.apply(
    lambda r: count_3x_spe_pokemon(r, "p1", "p2"), axis=1
)
p2_spe_counts = X_train.apply(
    lambda r: count_3x_spe_pokemon(r, "p2", "p1"), axis=1
)
X_train["spe_adv_diff"] = p2_spe_counts - p1_spe_counts
```

```python
# Update CV dataset with all three engineered features
X_cv = X_train[["hp_sum_diff", "type_adv_diff", "spe_adv_diff"]]

# Re-run the 5-Fold Stratified Cross-Validation
for name, model in models.items():
    scores = cross_val_score(model, X_cv, y_train, cv=cv, scoring="accuracy")
    print(f"{name:20} | Mean Accuracy: {scores.mean():.4f} (±{scores.std():.4f})")
```

```python
# Calculate the absolute fastest feature
p1_max_spe = X_train[[f"p1_mon{i}_spe" for i in range(1, 7)]].max(axis=1)
p2_max_spe = X_train[[f"p2_mon{i}_spe" for i in range(1, 7)]].max(axis=1)

X_train["p1_has_fastest_mon_on_field"] = (p1_max_spe > p2_max_spe).astype(int)

# Update CV dataset with all four engineered features
X_cv = X_train[
    [
        "hp_sum_diff",
        "type_adv_diff",
        "spe_adv_diff",
        "p1_has_fastest_mon_on_field",
    ]
]

# Re-run the 5-Fold Stratified Cross-Validation
for name, model in models.items():
    scores = cross_val_score(model, X_cv, y_train, cv=cv, scoring="accuracy")
    print(f"{name:20} | Mean Accuracy: {scores.mean():.4f} (±{scores.std():.4f})")
```
