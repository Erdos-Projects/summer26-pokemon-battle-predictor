```python
# Some basic imports
import numpy as np
import pandas as pd

import json, re, time
import logging, requests
import itertools

from pathlib import Path
from typing import Optional
from dataclasses import dataclass, asdict
from urllib.parse import urlencode, urljoin

# Useful 'constants'
REPLAY_DIR = Path('./replays/gen9-randombattle')
USER_DIR = Path('./users')
```

```python
ratings_data = []

for item in USER_DIR.iterdir() : 
    if (item.is_file() and item.name[0]!='.') : 
        with open(item, 'r') as file: 
            user_json = json.load(file)
        try:
            rating = user_json.get('ratings',{})
            ratings_data.append(rating.get('gen9randombattle',0))
        except:
            print("Could not get `gen9randombattle` rating for %s" % item.stem)
            continue
    elif (item.name[0]=='.') : continue
    else : print("invalid file"); continue
```
